## Anatomy of a Matplotlib plot - Workflow for matplotlib

```python
# 0. Import matplotlib - setup
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot the data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple plot",
       ylabel='Y axis',
       xlabel='X label')

# 5. Save figure
fig.savefig('/Users/tamir_gez/Documents/PythonProjects/MachineLearningZTM/Images/sample-plot.png')
```

## Two options to subplots

### Option 1 -

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(8,5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(num_butter_prices.keys(), num_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

### Option 2 -

```python
fig, ax =  plt.subplots(nrows=2,
                        ncols=2,
                        figsize=(8,5))

ax[0,0].plot(x, x*2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(num_butter_prices.keys(), num_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot.hist();
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
```

```python
fig, ax = plt.subplots(figsize=(10,8))

over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

ax.set_xlim([45, 100])
```

```python
fig, ax = plt.subplots(figsize=(10,8))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target']);

ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), ncols=1, nrows=2,
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'])

ax0.set(title='Heart Disease And Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholesterol')

# Add legend for the plot
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(over_50['chol'].mean(), linestyle='--')

scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'])

ax1.set(title='Heart Disease And Max Heart Rate',
        xlabel='Age',
        ylabel='Thal')

# Add legend for the plot
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle('Heart Disease Analaysis', fontsize=24, fontweight='bold')
fig.tight_layout()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(10,8))

scatter = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter');

ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol')

ax.legend(*scatter.legend_elements(), title="Target")

ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10), ncols=1, nrows=2,
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')

ax0.set(title='Heart Disease And Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholesterol')
ax0.set_xlim([50, 80])

# Add legend for the plot
ax0.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax0.axhline(over_50['chol'].mean(), linestyle='--')

scatter1 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                       cmap='winter')

ax1.set(title='Heart Disease And Max Heart Rate',
        xlabel='Age',
        ylabel='Max Heart Rate')
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add legend for the plot
ax1.legend(*scatter.legend_elements(), title='Target')

# Add a meanline
ax1.axhline(over_50['thalach'].mean(), linestyle='--')

fig.suptitle('Heart Disease Analaysis', fontsize=24, fontweight='bold')
fig.tight_layout()
```
